```python
print('******WELCOME TO TIMES BANK******')
print('  ')
while True:
    print(' Please Select One Option to Continue ')
    print(' ')
    print('1.Open a new account')
    print('2.Withdraw Money')
    print('3.Deposit Money')
    print('4.Check Customers and balance')
    print('5.Exit/quit')
    print(' ')
    print(' ')
    pin='1234'
    i=int(input('please enter your choice:'))
    bal=200000

    if i==1:
        a=input('enter your name :')
        g=input('enter your fathers name :')
        z=input('enter your DOB :')
        c=input('enter your address :')
        d=input('enter your aadhar number :')
        e=input('select your account type: \n      Savings \n      Current \nAccount Type:')
        f=input('enter your mobile number :')
        if len(f)==10:
            print('*******************Please Check Your Entered Details:***********************')
            print('Name            :',a)
            print("Father's Name   :",g)
            print('DOB             :',z)
            print('Address         :',c)
            print('Aadhar          :',d)
            print('Account Type    :',e)
            print('Mobile Number   :',f)
            p=input('IF Entered Details Are Correct Please Type yes to Proceed:')
            if p=='yes':
                print('**************Your Account Has Been Created Succesfully****************')
            else:
                print('Please Re-enter Your Details')
        else:
            print('You have entered only',len(f),'digit please enter correct mobile number')

    elif i==2:
        a=input('please select your account type : \n      Savings \n      Current \nAccount Type:')
        b=input('Enter your 12 digit account number :')
        if len(b)==12:
            c=int(input('Enter the amount to withdraw :'))
            if c>200000:
                print('*' * 24)
                print(' Insufficient Balance ')
                print('*' * 24)
            elif c<=200000:
                d=input('Enter your pin code :')
                if d==pin:
                    print('***************your amount',c,'has been withdraw succesfully**************')
                    print(' ')
                    print(' ')
                    print('*' * 40)
                    print('Available Balance:',bal-c)
                    print('*' * 40) 
                else:
                    print('You Have Entered Wrong Pin')
        else:
            print('You Have Entered Only',len(b),'numbers .please type your account number properly')
    elif i==3:
        a=input('Please enter your 12 digit account number :')
        if len(a)==12:
            b=int(input('Enter the amount to be deposited :'))
            c=input('Enter your pin :')
            if c==pin:
                print('*************Your amount has been deposited succesfully***************')
                print('*' * 40)
                print('     Available Balance:',bal+b)
                print('*' * 40) 
            else:
                print('Please Enter Correct Pin Number:')
        else:
            print('You Have Entered Only',len(a),'numbers')
    elif i==4:
        a=input('Select your account type : \n      Savings \n      Current \nAccount Type:')
        b=input('Enter your account number :')
        c=input('Enter your pin :')
        if c==pin:
            print('*' * 40)
            print('      Your current account balance is:',bal)
            print('*' * 40) 
        else:
            print('You Have Entered A Wrong PIN')
    elif i==5:
        print('EXIT SUCCESFULLY')
        break
        
    else:
        print('*' *30)
        print(' Please Select Correct Option')
        print('*' *30)
```
